        Making all values in column Golden Globe - Drama a string value

```python
picture["Golden Globe - Drama"] = picture["Golden Globe - Drama"].astype(str)
picture
```

```python
ids = picture['Golden Globe - Drama']
for i, id_ in enumerate(ids):
    print(f"id {i}: {id_}\t{id_.__class__.__name__}")
```
